```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import  precision_score, recall_score, f1_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
th='/content/Covid-19 Twitter Dataset (Apr-Jun 2020).csv'
z=pd.read_csv(path, encoding='latin-1')
df=pd.DataFrame(z)
df
```

```python
df.isnull().sum()
```

```python
df1=df.dropna()
df1.isnull().sum()
df1
```

```python
df1.isnull().sum()
```

```python
X=df1['clean_tweet']
Y=df1['sentiment']
X
```

```python
Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
#Naive bayes Multinomial Classification
model =MultinomialNB(alpha=0.5)
model.fit(X_train_tfidf, y_train)
```

```python
y_pred_nb=model.predict(X_test_tfidf)
print(y_pred_nb)
accuracy_nb = accuracy_score(y_test, y_pred_nb)
precision_nb = precision_score(y_test, y_pred_nb, average='weighted')
recall_nb = recall_score(y_test, y_pred_nb, average='weighted')
f1_nb = f1_score(y_test, y_pred_nb, average='weighted')
print(f'Accuracy: {accuracy_nb}')
```

```python
report = classification_report(y_test, y_pred_nb)
print(report)
```

```python
#Support Vector Classification
model1 =SVC()
model1.fit(X_train_tfidf, y_train)
y_pred_svc=model1.predict(X_test_tfidf)
print(y_pred_svc)
accuracy_svc = accuracy_score(y_test, y_pred_svc)
precision_svc = precision_score(y_test, y_pred_svc, average='weighted')
recall_svc = recall_score(y_test, y_pred_svc, average='weighted')
f1_svc = f1_score(y_test, y_pred_svc, average='weighted')
print(f'Accuracy: {accuracy_svc}')
report = classification_report(y_test, y_pred_svc)
print(report)
```

```python
#Visualization
cm = confusion_matrix(y_test, y_pred_svc)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for SVC')
plt.show()

cm2 = confusion_matrix(y_test, y_pred_nb)
plt.figure(figsize=(8, 6))
sns.heatmap(cm2, annot=True, fmt="d", cmap="Greens")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for Naive Bayes')
plt.show()

```

```python
labels = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
svc_metrics = [accuracy_svc, precision_svc, recall_svc, f1_svc]
nb_metrics = [accuracy_nb, precision_nb, recall_nb, f1_nb]
width = 0.35  # Width of the bars

x = np.arange(len(labels))
fig, ax = plt.subplots(figsize=(10, 6))
bar1 = ax.bar(x - width/2, svc_metrics, width, label='SVC', color='b', alpha=0.7)
bar2 = ax.bar(x + width/2, nb_metrics, width, label='Naive Bayes', color='g', alpha=0.7)

#labels, title, and legend
ax.set_xlabel('Metrics')
ax.set_ylabel('Scores')
ax.set_title('Sentiment Analysis Model Comparison')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Displaying the chart
plt.tight_layout()
plt.show()
```
